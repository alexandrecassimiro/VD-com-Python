```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# size são quantos elementos haverá no vetor
eixo_y = np.random.randint(low = 100, high = 5000, size =50)
```

```python
eixo_y  
```

```python
# inserir o range de data
eixo_x = pd.date_range(start='2021-1-1', freq='D', periods= 50 )
```

```python
eixo_x
```

**Com matplotlib**

```python
# %matplotlib notebook deveria exibir o gráfico interativo. mas não ocorre
%matplotlib inline

plt.figure(figsize  = (8,5))
plt.plot(eixo_x,eixo_y)
plt.title('Meu primeiro Linhas')
plt.show()
```

```python
# criar uma figura com o gráfico dinâmico
fig = px.line(x = eixo_x, y = eixo_y, title='Fig interativa')
fig.show()
```
